Fido permite la flexibilidad de buscar datos de múltiples instrumentos, longitudes de onda, tiempos, etc., incluso cuando los datos se obtienen a través de diferentes clientes.

utilizando | : une consultas al igual que el operador OR.

buscando datos GOES/XRS y Fermi/GBM durante un período de tiempo particular.

```python
result = Fido.search(a.Time("2013-10-28 01:30", "2013-10-28 02:00"),
                     a.Instrument.xrs  | a.Instrument.gbm)
```

```python
result
```

```python
result_HMI = Fido.search(a.Time("2022-12-10 11:50", "2022-12-10 12:25"),
                     a.Instrument.hmi  | a.Instrument.gbm)
```

```python
result_HMI
```

se puede especificar más la búsqueda para cada instrumento,

 para buscar solo datos del satélite 15 del GOES y del detector n5 para GBM, esto se usa &

```python
result = Fido.search(a.Time("2013-10-28 01:00", "2013-10-28 03:00"),
                     a.Instrument.xrs & a.goes.SatelliteNumber(15) | a.Instrument.gbm & a.Detector("n5"))
```
